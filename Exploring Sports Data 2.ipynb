Below I randomly make matchups and wins to calculate accuracy on.

```python
teams = df9.Team.unique()
print teams
predict = []
x=0
while x<30:
    l=list(np.random.choice(teams, 2, replace=False))
    l.append(l[np.random.random_integers(0,1)])
    predict.append(l)
    
    x=x+1
    
predict = pd.DataFrame(predict)
predict.dropna(inplace=True)
predict.reset_index(inplace=True, drop=True)
predict.columns=['Team 1', 'Team 2', 'Winner']
print predict
```

Using total game time and individual player ratings I generate a weighted rating for each player. So, this is assuming for a given game I know how much time they will play and/or know how much time they DID play and will use that to predict the overall rating to determine the win.

```python
totalgametime = 90 #would need actual game time to get the weighted rating
df9['WeightedRating'] = df9['Rt']/(totalgametime/df9['MinutesPlayed'])
df9.head() #rating per player based on time in game
```

```python
rank9 = df9.groupby('Team').sum() #rating per team based on player's weighted ratings
rank9.sort(columns='WeightedRating', ascending=False) #sort them for fun
```

```python
#decide the winner in a given matchup
def whowon(lu, team1, team2):
    try:
        team1rate = lu.loc[team1, 'WeightedRating']
    except:
        return "No data"
    try:
        team2rate = lu.loc[team2, 'WeightedRating']
    except:
        return "No data"
    if team1rate > team2rate:
        return team1
    elif team1rate < team2rate:
        return team2
    elif team2rate == team1rate:
        return "Tie"
    else:
        return "Cannot compute."
```

```python
#predict for all of the fake matchups I made
i=0
winners = []
for r in predict.itertuples():
    winners.append(whowon(rank9, r[1], r[2]))
    i = i+1
predict['Calculated']=winners
```

We can calculate accuracy by doing line comparisons of calculated to actual wins and dividing by total predictions.

```python
def get_accuracy(calc, actual):
    accuracy = []
    for i in xrange(len(calc)):
        if calc[i] == actual[i]:
            accuracy.append(1.0)
        else:
            accuracy.append(0.0)
    accuracy = sum(accuracy) / len(accuracy)
    print accuracy*100.0,'%'
```

```python
get_accuracy(predict.Calculated, predict.Winner)
```
