Wrangling some data so it is nice to us.

```python
df9 = df9.replace('-',0)
```

```python
df9[['Goal','A','Yel','Red','SpG','PS%','AW','MoM','Rt']]=df9[['Goal','A','Yel','Red','SpG','PS%','AW','MoM','Rt']].astype(float)
```

```python
df9['SumStat'] = df9.ix[:,6:14].sum(axis=1, numeric_only=True)
```

Arbitrarily added up all of the stats available to me per player to group on per team to get a team stat. A later iteration seeks to weight these by minutes per player.

```python
df9.head()
```

```python
rank9 = df9.groupby("Team")['SumStat'].sum()
```

```python
rank9.sort(ascending=False) #sort them 
```

Adding up player stats per team yield the team stat which we can use to rank the teams by a unitless likelihood of win rate in a given season.

```python
rank9
```

These arbitrary stat numbers do not convert into wins per season and we have no defined way to convert them. We could turn the stat numbers into ratios by dividing them against the lowest performing team and assign wins based on these ratios from the number of games played per team. I make the assumption here that the team with the lowest aggregate stats will win none of its games.

```python
rank9 = pd.DataFrame(rank9)
```

```python
rank9
```

```python
rank9['StatRatio'] = rank9/rank9.ix[rank9.shape[0]-1]-1
```

```python
rank9['StatRatio'] #Here are the ratios.
```

```python
rank9['CalculatedWins'] = rank9['StatRatio']*38
```

```python
rank9['CalculatedWins'] = rank9['CalculatedWins'].astype(int)
```

```python
rank9
```

```python
final = pd.merge(rank9, predict, left_index=True, right_on='Team')
```

```python
final.reset_index(inplace=True)
```

```python
final
```

We can calculate accuracy by doing line comparisons of calculated to actual wins and dividing by total predictions.

```python
def get_accuracy(calc, actual):
    accuracy = []
    for i in xrange(len(calc)):
        if int(calc[i]) == int(actual[i]):
            accuracy.append(1.0)
        else:
            accuracy.append(0.0)
    accuracy = sum(accuracy) / len(accuracy)
    print accuracy*100.0,'%'
```

```python
get_accuracy(final.CalculatedWins, final.W)
```
